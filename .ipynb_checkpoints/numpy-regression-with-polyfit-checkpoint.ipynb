```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr = np.array([1, 1, 2, 3, 4, 6, 3, 4, 7, 5, 8, 10, 7, 9])
xs = np.arange(1,len(arr) + 1)  # 14 pts along the x-axis
m, b = np.polyfit(xs, arr, 1)  # 3rd argument specifies degree of best fit, if you want quadratic, pass 2, cubic pass 3
```

```python
plt.plot(xs, arr, 'b.')
plt.plot(xs, m*xs + b, 'r')  # think y = mx + b
```

#### Now consider some non-linear data

```python
y = np.array([0.0, 0.8, 0.9, 0.1, -0.8, -1.0])
```

```python
x = np.arange(1,len(y)+1)  # one x for each y
```

```python
second = np.polyfit(x, y, 2)  # fit 2nd degree polynomial, returns polynomial coefficients, highest power first
```

```python
third = np.polyfit(x, y, 3)  # fit a 3rd degree polynomial
```

```python
p2 = np.poly1d(second)  # defines p2 as the polynomial function of the best fit found by polyfit for second
```

```python
p3 = np.poly1d(third)  # defines p3 as the polynomial function of the best fit found by polyfit for third
```

```python
xp = np.linspace(0, 6, 100)  #  need more xs to make the curve plots smooth
```

```python
plt.plot(x, y, 'b.', xp, p2(xp), 'r', xp, p3(xp), 'g')
```
