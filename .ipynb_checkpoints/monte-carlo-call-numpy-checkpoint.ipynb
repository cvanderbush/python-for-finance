```python
import math
import numpy as np
from time import time
import matplotlib.pyplot as plt
```

```python
np.random.seed(20000)
```

```python
S0 = 100.0
K = 105.0
T = 1.0
r = 0.05
sigma = 0.2
M = 50
dt = T / M
I = 250000
```

```python
# each row one of M time steps, each column one of I different stock paths
S = np.zeros((M + 1, I))
```

```python
S[0] = S0
```

```python
np.shape(S)  # S0 plus 50 timesteps
```

```python
t0 = time()
for t in range(1, M + 1):
    z = np.random.standard_normal(I)  # vector of I random numbers
    S[t] = S[t-1] * np.exp((r - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * z) # vector of I prices at next timestep
C0 = np.exp(-r * T) * np.sum(np.maximum(S[-1] - K, 0)) / I

tnp = time() - t0
print("Value of call is {:.4f}".format(C0))
print("Calculation time is {:.2f} seconds".format(tnp))
```

#### Plot distribution of final path state

```python
plt.grid(1, zorder=1)
plt.hist(S[-1], bins = 50, zorder=2)
```

#### Plot first 10 price paths

```python
plt.figure(figsize = (12, 9))
plt.plot(S[:, :10])
plt.grid(True)
plt.axis([0, 50, 70, 150])
plt.xlabel('time periods')
plt.ylabel('price')
plt.title('Price Paths over 50 time periods')
```
